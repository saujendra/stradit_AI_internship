```python
#3 block of code

def count_lines(path):

    try:

        file = open(path, "r")


        count = 0


        for line in file:

            if line.strip() != "":

                count += 1


        file.close()

        return count


    except FileNotFoundError:

        return 0


print("Total lines:",count_lines("demo.txt"))
```

```python
#4
def word_count(path):
    with open(path, "r") as file:
            text = file.read()
            words = text.split()
            file.seek(0)
            print(file.read())
            file.close()
            
    return len(words)


print("total word count:",word_count("demo.txt"))
```

```python
#5
def tail(path, n):

    file = open(path, "r")


    lines = file.readlines()


    file.close()


    return lines[-n:]


print(tail("data.txt", 2))
```

```python
def replace_in_file(path, old, new):


    with open(path, "r") as file:

        text = file.read()


    text = text.replace(old, new)


    with open(path, "w") as file:

        file.write(text)


replace_in_file("data.txt", "Python", "Java")


print("Replacement Done!")
```

```python
import csv


with open("people.csv", "r") as file:


    reader = csv.DictReader(file)


    adults = []


    for row in reader:


        if int(row["age"]) >= 30:

            adults.append(row)


with open("adults.csv", "w", newline="") as file:


    fieldnames = ["name", "age", "city"]


    writer = csv.DictWriter(file, fieldnames=fieldnames)


    writer.writeheader()


    writer.writerows(adults)


print("adults.csv created")
```

```python
import csv


def csv_to_dict_list(path):


    with open(path, "r") as file:


        reader = csv.DictReader(file)


        data = []


        for row in reader:

            data.append(row)


    return data


print(csv_to_dict_list("people.csv"))
```

```python
import csv


def dict_list_to_csv(records, path):


    with open(path, "w", newline="") as file:


        fieldnames = records[0].keys()


        writer = csv.DictWriter(file, fieldnames=fieldnames)


        writer.writeheader()


        writer.writerows(records)


data = [

    {"name": "Rahul", "age": 25, "city": "Pune"},

    {"name": "Aman", "age": 30, "city": "Delhi"}

]


dict_list_to_csv(data, "new_people.csv")


print("CSV Written Successfully")
```

```python
# 1. Write a safe_int(s, default=0) that tries int(s) and returns default on 
# ValueError
def safe_int(s, default=0):


    try:

        return int(s)


    except ValueError:

        return default


print(safe_int("10"))

print(safe_int("abc"))
```

```python
#  Write a function divide(a, b) that handles both TypeError (non-numbers) and 
# ZeroDivisionError, returning None for invalid input.

def divide(a, b):

    try:
        return a / b

    except ZeroDivisionError:
        print("Cannot divide by zero")
        return None

    except TypeError:
        print("Invalid input")
        return None


print(divide(10, 2))
print(divide(10, 0))
print(divide("a", 2))
```

```python

# 4. Write a function that reads numbers from a file (one per line). Skip any line that isn't a valid 
# number, but count and report how many were skipped


skipped = 0


with open("numbers.txt", "r") as file:

    for line in file:

        try:
            number = int(line.strip())

            print(number)

        except ValueError:

            skipped += 1


print("Skipped lines =", skipped)
```

```python
# Create a custom exception EmptyFileError. Write a function that reads a file and raises 
# this exception if the file is empty.
class EmptyFileError(Exception):
    pass


def check_file(path):

    with open(path, "r") as file:

        data = file.read()

        if data == "":
            raise EmptyFileError("File is empty")

        print(data)


try:
    check_file("data.txt")

except EmptyFileError as e:
    print(e)
```

```python
import csv


try:


    with open("students.csv", "r") as file:


        reader = csv.DictReader(file)


        result = []


        for row in reader:


            total = 0


            for subject in ["math", "science", "english"]:


                try:

                    marks = int(row[subject])


                except ValueError:


                    print(f"Warning: Invalid value in {row['name']} -> {subject}")


                    marks = 0


                total += marks


            average = total / 3


            row["total"] = total

            row["average"] = average


            result.append(row)


    with open("students_with_grades.csv", "w", newline="") as file:


        fieldnames = [

            "name",

            "math",

            "science",

            "english",

            "total",

            "average"

        ]


        writer = csv.DictWriter(file, fieldnames=fieldnames)


        writer.writeheader()


        writer.writerows(result)


    print("students_with_grades.csv created")


except FileNotFoundError:


    print("students.csv file not found")
```
