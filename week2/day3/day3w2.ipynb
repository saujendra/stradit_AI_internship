```python
# 3. For a 1-D array a = np.array([5, 10, 15, 20, 25]), standardise it: (a - 
# mean) / std. Verify the result has mean 0 and std 1.
a = np.array([5, 10, 15, 20, 25])
mean=np.mean(a)
sd=np.std(a)

stad=(a-mean)/sd

print(stad)
print("\nVerification process:\n")

if np.isclose(np.mean(stad), 0) and np.isclose(np.std(stad), 1):
    print("Mean = 0 and Std = 1")
else:
    print("Verification failed")



```
