```python
# . What is the average fare paid by passengers who survived vs those who did not? (Hint: 
# groupby("Survived")["Fare"].mean()
print("avg paid by pasenger in  survived :\n",df.groupby("Survived")["Fare"].mean())
 #Average age per passenger class.
print("Average age per passengerclass:\n",df.groupby("Pclass")["Age"].mean())
#Median age per (Pclass, Sex) combination.
print("Median age per passenger (Pclass, Sex) )class:\n",df.groupby(["Pclass","Sex"])["Age"].median())
# Survival rate per embarkation port.
print("survival rate per embarkation rate port:",df.groupby("Embarked")["Survived"].mean())
#Maximum fare per class
print("Maximum Fare class:\n",df.groupby("Pclass")["Fare"].max())
```

```python
# Block E — Cleaning & Engineering (4 tasks)

# Fill missing Age with the median age. Fill missing Embarked with the most common port. 
# Drop the Cabin column.

median_age=df["Age"].median()
print(df["Age"])
print("Missing ages now:",df["Age"].isna().sum())
df["Age"]=df["Age"].fillna(median_age)
print(df["Age"])
print("Missing ages now:",df["Age"].isna().sum())

mode_erb=df["Embarked"].mode()
print("Missing emberked now:",df["Embarked"].isna().sum())
df["Age"]=df["Age"].fillna(mode_erb)
print(df["Embarked"])

print(df.head())

```
