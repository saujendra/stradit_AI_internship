```python
import pandas as pd
df = pd.read_csv("titanic 1.1.csv")
print("Dataset loaded successfully!\n")


```

```python

# Clean missing values and drop Cabin
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df = df.drop("Cabin", axis=1)

# Create Title, FamilySize, IsAlone, FareCategory
df["Title"] = df["Name"].str.split(", ").str[1].str.split(".").str[0]
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df["IsAlone"] = (df["FamilySize"] == 1).astype(int)

df["FareCategory"] = pd.cut(
    df["Fare"],
    bins=[0, 10, 50, 100, 600],
    labels=["low", "mid", "high", "premium"]
)

# Show survival rates
print("\nSurvival Rate by Title:\n")
print(df.groupby("Title")["Survived"].mean())

print("\nSurvival Rate by FamilySize:\n")
print(df.groupby("FamilySize")["Survived"].mean())

print("\nSurvival Rate by FareCategory:\n")
print(df.groupby("FareCategory")["Survived"].mean())

# Save cleaned dataset
df.to_csv("titanic_clean.csv", index=False)

print("\nCleaned dataset saved successfully!")

```
