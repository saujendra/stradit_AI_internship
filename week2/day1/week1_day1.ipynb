```python
#  Given transactions = [{"user":"a","amt":10},
# {"user""b","amt":7},{"user":"a","amt":3}], build a dict mapping each 
transactions = [
    {"user": "a", "amt": 10},
    {"user": "b", "amt": 7},
    {"user": "a", "amt": 3}
]

user_totals = {}

for i in transactions:
    user = i["user"]
    amt = i["amt"]
    print(f"user={user} amount={amt}")
    user_totals[user] = user_totals.get(user, 0) + amt

print(user_totals)

    
```
